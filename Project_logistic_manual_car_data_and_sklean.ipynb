# **Logistic**

```python
import numpy as np
import pandas as pd

df = pd.read_csv('/content/car_data_(task_classification).csv')
df
```

```python
df.info()
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].astype('category').cat.codes

X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values.reshape(-1, 1)
```

```python
X[:5]
```

```python
y[:5]
```

```python
# Normalization
X = (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
X.shape
```

```python
n_samples, n_features = X.shape
W = np.zeros((n_features, 1))
b = 0.0
alpha = 0.1
epochs = 2000
```

```python
# def sigmoid
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def get_accuracy(y_true, y_pred):
    y_pred_cls = (y_pred >= 0.5).astype(int)
    return np.mean(y_pred_cls == y_true)
```

```python
# Training
for epoch in range(epochs):
    z = np.dot(X, W) + b
    y_pred = sigmoid(z)

    error = y_pred - y

    dW = np.dot(X.T, error) / n_samples
    db = np.sum(error) / n_samples

    W -= alpha * dW
    b -= alpha * db

    if epoch % 200 == 0:
     acc = get_accuracy(y, y_pred)
     print(f"Epoch {epoch}  - Accuracy: {acc:.4f}")
```

```python
# Accuracy Score
y_pred_final = sigmoid(np.dot(X, W) + b)
y_pred_class = (y_pred_final >= 0.5).astype(int)
accuracy = get_accuracy(y, y_pred_final)
print(" Accuracy:", round(accuracy * 100, 2))
```

```python
# MSE
mse = np.mean((y - y_pred_class)**2)
print("Mean Squared Error (MSE):", round(mse, 4))
```

# **Sklearn**

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, mean_squared_error

df = pd.read_csv('/content/car_data_(task_classification).csv')

for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].astype('category').cat.codes

X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values.reshape(-1, 1)

model = LogisticRegression()
model.fit(X, y)

y_pred_prob = model.predict_proba(X)[:, 1]
y_pred_cls = (y_pred_prob >= 0.5).astype(int)

acc = accuracy_score(y, y_pred_cls)
mse = mean_squared_error(y, y_pred_cls)

print(f"Accuracy: {acc:.4f}")
print(f"MSE: {mse:.4f}")
```
