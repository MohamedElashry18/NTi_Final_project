```python
sum_x = sum(X)
sum_y = sum(y)
sum_xy = sum([X[i] * y[i] for i in range(n)])
sum_x2 = sum([X[i] ** 2 for i in range(n)])
```

```python
w = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2) # b
b = (sum_y - w * sum_x) / n # a
```

```python
y_pred = [w * X[i] + b for i in range(n)]
```

```python
mse = sum([(y[i] - y_pred[i]) ** 2 for i in range(n)]) / n
```

```python
mean_y = sum_y / n
ss_tot = sum([(y[i] - mean_y) ** 2 for i in range(n)])
ss_res = sum([(y[i] - y_pred[i]) ** 2 for i in range(n)])
r2 = 1 - (ss_res / ss_tot)
```

```python
print(f"y = W : {w:.2f} X + b : {b:.2f}")
print(f"MSE: {mse:.2f}")
print(f"R: {r2:.4f}\n")
```
